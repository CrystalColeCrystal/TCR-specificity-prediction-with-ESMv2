### 2. Import  libraries

```python
import pandas as pd
import evaluate
import numpy as np
import torch

from transformers import AutoTokenizer
from datasets import Dataset
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer, AutoModel

from sklearn import metrics
from scipy.special import softmax
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve, auc
```

### 3. Load the model from checkpoint in main/weights/ of the current repo

```python
model_checkpoint = "esm2_t6_8M_UR50D_300epoch"
```

```python
model = AutoModelForSequenceClassification.from_pretrained(model_checkpoint, local_files_only=True)
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint, local_files_only=True)
```

### 4. Load the data for testing from main/data/

```python
def dataset_from_csv(file, tokenizer):
    
    ''' A utility function'''
    
    data = pd.read_csv(file)
    data_sequences = list(data['cdr3.alpha']+data['cdr3.beta']+data['antigen.epitope'])
    data_labels = list(data['target'])
    data_tokenized = tokenizer(data_sequences)
    dataset = Dataset.from_dict(data_tokenized)
    dataset = dataset.add_column("labels", data_labels)
    return dataset
```

```python
hard_test = dataset_from_csv('~/TCR-specificity-prediction-with-ESMv2/data/hard_test.csv', tokenizer)
```

```python
train_dataset = dataset_from_csv('~/TCR-specificity-prediction-with-ESMv2/data/train.csv', tokenizer)
easy_val = dataset_from_csv('~/TCR-specificity-prediction-with-ESMv2/data/easy_val.csv', tokenizer)
```

```python
model_name = model_checkpoint.split("/")[-1]
batch_size = 2048

args = TrainingArguments(
    f"{model_name}-batch-size-test-finetuned-localization",
    evaluation_strategy = "epoch",
    save_strategy = "epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    num_train_epochs=300,
    weight_decay=0.01,
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
    push_to_hub=False,
    label_names=["labels"],
    report_to=["comet_ml"]
)
```

```python
clf_metrics = evaluate.combine(["accuracy", "f1", "precision", "recall"])

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return clf_metrics.compute(predictions=predictions, references=labels)
```

```python
trainer = Trainer(
    model,
    args,
    train_dataset=train_dataset,
    eval_dataset=easy_val,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

### 6. Calculate ROC-AUc and build the curve

Easy subset:

```python
roc_auc_score = evaluate.load("roc_auc")
refs = easy_val["labels"]
pred_scores = softmax(trainer.predict(test_dataset=easy_val)[0][:,1])
results = roc_auc_score.compute(references=refs, prediction_scores=pred_scores)
print(round(results['roc_auc'], 2))
```

```python
fpr, tpr, thresholds = metrics.roc_curve(refs, pred_scores, pos_label=1)

plot = plt.plot(fpr, tpr, lw = 2, label = 'ROC AUC: {:.2f}'.format(results['roc_auc']))
plt.plot([0, 1], [0, 1],
         linestyle = '--',
         color = (0.6, 0.6, 0.6),
         label = 'random guessing')
plt.plot([0, 0, 1], [0, 1, 1],
         linestyle = ':',
         color = 'black', 
         label = 'perfect performance')

plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.title('Receiver Operator Characteristic')
plt.legend(loc = "lower right")
plt.tight_layout()
fig = plot[0].get_figure()
fig.savefig("ROC.svg")
plt.show()
```
